## depression analysis

```python
depressed_students = df[
    (df['Working Professional or Student'] == 'Student') & 
    (df['Depression'] == 1)
     ]

print("\nDepressed students count:", len(depressed_students))
```

```python
plt.figure(figsize=(8,5))
sns.countplot(
    x='Working Professional or Student', 
    hue='Depression', 
    data=df,
    palette='Set2'
)
plt.title("4. Depression: Students vs Professionals")
plt.show()
```

```python
df.head()
```

```python
depressed_counts = (df[df['Working Professional or Student'] == 'Working Professional'].groupby('Profession')['Depression'].sum() .sort_values(ascending=False))
print(depressed_counts.head(10))
```

# top 10 Depressed profession

```python
plt.figure(figsize=(10,5))
sns.barplot(
    x=depressed_counts.head(10).index, 
    y=depressed_counts.head(10).values,
    palette="magma"
)
plt.title("Top 10 Depressed Professions")
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
suicidal_counts = (
    df[df['Working Professional or Student'] == 'Working Professional']
    .groupby(['Profession', 'Have you ever had suicidal thoughts ?'])
    .size()
    .unstack()
)
print(suicidal_counts)
```

 # Professions with Highest Suicidal Thoughts 

```python

!pip install squarify
import squarify 

top_20 = suicidal_counts.sort_values('Yes', ascending=False).head(20)
plt.figure(figsize=(12, 8))
squarify.plot(
    sizes=top_20['Yes'],
    label=[f"{prof}\n{count}" for prof, count in zip(top_20.index, top_20['Yes'])],
    color=sns.color_palette('coolwarm_r', len(top_20)),
    alpha=0.7,
    text_kwargs={'fontsize':9}
)
plt.title('Professions with Highest Suicidal Thoughts ', pad=20)
plt.axis('off')
plt.tight_layout()
plt.show()
```

 # Most of the depressed people in the dataset are sleeping less than 5 hours.
The number drops as sleep hours increase, which clearly shows that lack of sleep is strongly linked with depression.
Proper sleep might be an important factor for mental health, especially for people at risk of depression.
